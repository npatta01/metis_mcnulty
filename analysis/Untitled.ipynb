```python
def get_classification_result(filename):
    files = {'image_list': open(filename, 'rb')}
    payload = {'job_id':job_id}
    r=requests.post(api_server,data=payload,files=files)
    classified = json.loads(r.content)['classifications']
    return classified
    
def parse_sub_file_results(classified):
    results={}
    for (key,value) in classified.items():
        most_accurate_label=value[0][0]
        just_file_name= key.split('/')[-1].replace('.png','')
        results[just_file_name]=most_accurate_label
        #results.append( (just_file_name, most_accurate_label ))
    return results

def get_most_confident_for_file(filename):
    classified=get_classification_result(filename)
    results = parse_sub_file_results(classified)
    return results

def get_results_for_files_in_folder(folder_name):
    results ={}
    for f in listdir(base_path):
        full_path=join(folder_name,f)
        sub_results=get_most_confident_for_file (full_path)
        results.update(sub_results)
        #results = results + sub_results
    return results
```
